```python
from transformers import pipeline
```

```python
spam_analyzer = pipeline('text-classification',model = 'mrm8488/bert-tiny-finetuned-sms-spam-detection')
```

```python
texts = [
    "You have been selected for a free vacation to Dubai! Call now to claim your reward.",
    "Hey, can you send me the notes from today's lecture?",
    "Act fast! Lowest insurance rates available only today. Apply now!",
    "Your OTP for login is 482913. Do not share it with anyone.",
    "Earn $5000 per week working from home. No experience required!",
    "Mom, I will be late for dinner. Don’t wait for me.",
    "Get cheap medicines without prescription. 100% guaranteed delivery.",
    "Meeting has been rescheduled to 3 PM. Please confirm your availability.",
    "You won a lottery of $2 million! Send your bank details immediately.",
    "Can you review my pull request when you get time?",
    "Exclusive deal just for you! 70% discount on branded shoes.",
    "Let’s catch up this weekend for coffee.",
    "Your parcel is out for delivery and will arrive today.",
    "Click here to reset your password or your account will be locked.",
    "Final reminder: Pay your electricity bill to avoid disconnection.",
    "Free entry in 2 a weekly competition to win FA Cup tickets. Text WIN to 87121 now!",
    "Please Share your OTP for lottery of $5000000"
]


results = spam_analyzer(texts)
classes = {'negative':'SPAM','positive':'Not SPAM','neutral':'Neutral'}
for result in results:
    # print(classes[result['label']],end=' ')
    # print(result['score'])
    print(result)
    

```
