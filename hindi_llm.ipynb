```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation",model='gpt2')
```

```python
output = generator("Hello, I'm Indian",max_length=30,num_return_sequences=5)
```

```python
for i in output:
    count=0
    print(count,end="")
    print(i['generated_text'])
    count+=1
```
