```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "sshleifer/distilbart-cnn-12-6"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = """Gateway Group of Companies is a global IT services and consulting organization that provides end-to-end technology solutions to enterprises worldwide. Founded in 1997, the company specializes in digital transformation, product engineering, enterprise mobility, cloud computing, DevOps, data analytics, AI/ML solutions, and enterprise application development. Gateway Group serves clients across industries including healthcare, fintech, manufacturing, logistics, education, and eCommerce. The organization focuses on delivering scalable, secure, and performance-driven solutions using modern technology stacks such as Python, .NET, Java, React, Angular, AWS, and Azure. With development centers in India and offices across multiple countries, the company emphasizes innovation, quality assurance, and customer-centric delivery models. Gateway follows agile methodologies to ensure faster time-to-market and transparent project execution. The company also invests in research and development to stay aligned with emerging technologies like artificial intelligence, blockchain, IoT, and automation. Through strategic partnerships and a skilled workforce, Gateway Group of Companies aims to help businesses modernize legacy systems, optimize operational efficiency, and achieve sustainable digital growth in a competitive global market.
"""

inputs = tokenizer.encode(
    "summarize: " + text,
    return_tensors="pt",
    max_length=512,
    truncation=True
)

summary_ids = model.generate(
    inputs,
    max_length=100,
    min_length=25,
    length_penalty=2.0,
    num_beams=4
)

summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```
